```python
def sum(n):
  sum=0;
  for i in range(n+1):
    sum+=i
  print(sum)
sum(10)
```

```python
def no(number):
  reversednumber=0
  while(number>0):
    remainder=number%10
    reversednumber=10*reversednumber+remainder
    number=number//10
  return reversednumber
number=no(123)
print(number)
```

```python
def count(number):
  reversednumber=0
  count=0
  while(number>0):
    remainder=number%10
    reversednumber=10*reversednumber+remainder
    number=number//10
    count+=1
  return count
number=count(123)
print(number)

```

```python
def palindrome(number):
  reversednumber=0
  original=number
  while(number>0):
    remainder=number%10
    reversednumber=10*reversednumber+remainder
    number=number//10
  if(original==reversednumber):
    return True
  else:
    return False
palindrome(12321)

```

```python
def fibonacci(number):
  a,b=0,1
  for i in range(number+1):
    print(a)
    a,b=b,a+b
fibonacci(10)
```

```python
def addition(a,b):
  print(a+b)
def subtraction(a,b):
  print(a-b)
def multiplication(a,b):
  print(a*b)
def division(a,b):
  print(a/b)

print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
operation=int(input("Enter the operation choice:"))
a=int(input("Enter number:"))
b=int(input("Enter second number"))
if(operation==1):
  addition(a,b)
elif(operation==2):
  subtraction(a,b)
elif(operation==3):
  multiplication(a,b)
else:
  division(a,b)

```
