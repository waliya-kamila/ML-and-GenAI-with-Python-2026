```python
# 1. Function to Print First 10 Natural Numbers
def print_natural():
    for i in range(1, 11):
        print(i, end=" ")
    print()

# 2. Function to Calculate Sum of First N Natural Numbers
def sum_natural(n):
    return n * (n + 1) // 2

# 3. Function to Reverse a Number
def reverse_number(num):
    return int(str(num)[::-1])

# 4. Function to Count Digits in a Number
def count_digits(num):
    return len(str(abs(num)))

# 5. Function to Check Palindrome Number
def is_palindrome(num):
    return str(num) == str(num)[::-1]

# 6. Function to Generate Fibonacci Series
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
    print()

# 7. Calculator Using Functions
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b):
    if b == 0:
        return "Error! Division by zero"
    return a / b

def calculator():
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    op = input("Enter operator (+, -, *, /): ")
    if op == "+": print("Result:", add(a, b))
    elif op == "-": print("Result:", subtract(a, b))
    elif op == "*": print("Result:", multiply(a, b))
    elif op == "/": print("Result:", divide(a, b))
    else: print("Invalid operator")

# 8. Create a Text File and Store Student Details
def write_student_file():
    with open("students.txt", "w") as f:
        n = int(input("How many students? "))
        for i in range(n):
            name = input("Enter name: ")
            marks = input("Enter marks: ")
            f.write(f"Name: {name}, Marks: {marks}\n")
    print("File saved successfully!")

# 9. Read Data from a File
def read_student_file():
    try:
        with open("students.txt", "r") as f:
            print("\n--- Student Data ---")
            print(f.read())
    except FileNotFoundError:
        print("File not found! Run option 8 first.")

# 10. Handle Division by Zero Using Exception Handling
def safe_divide():
    try:
        a = float(input("Enter numerator: "))
        b = float(input("Enter denominator: "))
        print("Result:", a / b)
    except ZeroDivisionError:
        print("Error! Cannot divide by zero.")
    except ValueError:
        print("Error! Please enter valid numbers.")

# 11. Student Class
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        total = sum(self.marks)
        percentage = (total / (len(self.marks) * 100)) * 100
        print(f"\nName: {self.name}")
        print(f"Marks: {self.marks}")
        print(f"Total: {total}")
        print(f"Percentage: {percentage:.2f}%")
        print("Result:", "PASS" if percentage >= 40 else "FAIL")

# ===== MAIN MENU =====
print("=" * 40)
print("        FUNCTION PROGRAMS MENU")
print("=" * 40)
print("1. Print First 10 Natural Numbers")
print("2. Sum of First N Natural Numbers")
print("3. Reverse a Number")
print("4. Count Digits in a Number")
print("5. Check Palindrome Number")
print("6. Fibonacci Series")
print("7. Calculator")
print("8. Write Student File")
print("9. Read Student File")
print("10. Division with Exception Handling")
print("11. Student Class Demo")
print("=" * 40)

choice = int(input("Enter your choice (1-11): "))

if choice == 1:
    print_natural()
elif choice == 2:
    n = int(input("Enter N: "))
    print("Sum:", sum_natural(n))
elif choice == 3:
    num = int(input("Enter a number: "))
    print("Reversed:", reverse_number(num))
elif choice == 4:
    num = int(input("Enter a number: "))
    print("Digits:", count_digits(num))
elif choice == 5:
    num = int(input("Enter a number: "))
    print("Palindrome:", is_palindrome(num))
elif choice == 6:
    n = int(input("How many terms? "))
    fibonacci(n)
elif choice == 7:
    calculator()
elif choice == 8:
    write_student_file()
elif choice == 9:
    read_student_file()
elif choice == 10:
    safe_divide()
elif choice == 11:
    name = input("Enter student name: ")
    n = int(input("How many subjects? "))
    marks = []
    for i in range(n):
        m = float(input(f"Marks in subject {i+1}: "))
        marks.append(m)
    s = Student(name, marks)
    s.display()
else:
    print("Invalid choice!")
```
