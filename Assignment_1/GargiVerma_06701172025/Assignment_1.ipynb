```python
# Question 7: Student Report Program

name = "Gargi Verma"

m1 = 85
m2 = 90
m3 = 78
m4 = 88
m5 = 92

total = m1 + m2 + m3 + m4 + m5
percentage = (total / 500) * 100

print("Student Name:", name)
print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 90:
    print("Grade: A")
elif percentage >= 80:
    print("Grade: B")
elif percentage >= 70:
    print("Grade: C")
elif percentage >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```
