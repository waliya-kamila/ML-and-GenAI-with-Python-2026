```python
#1.print first 10 natural numbers
def print_natural_numbers():
    for i in range(1, 11):
        print(i)

print_natural_numbers()
```

```python
#2.sum of first N natural numbers
def sum_natural_numbers(n):
    return n * (n + 1) // 2

n = int(input("Enter N: "))
print("Sum =", sum_natural_numbers(n))
```

```python
#3.reverse a number
def reverse_number(num):
    rev = 0
    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num //= 10
    return rev

num = int(input("Enter a number: "))
print("Reversed number:", reverse_number(num))
```

```python
#4.count digits in a number
def count_digits(num):
    count = 0

    if num == 0:
        return 1

    while num > 0:
        count += 1
        num //= 10

    return count

num = int(input("Enter a number: "))
print("Number of digits:", count_digits(num))
```

```python
#5.check palindrome number
def is_palindrome(num):
    original = num
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num //= 10

    return original == reverse

num = int(input("Enter a number: "))

if is_palindrome(num):
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
#6.generate fibonacci series
def fibonacci(n):
    a, b = 0, 1

    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

n = int(input("How many terms? "))
fibonacci(n)
```

```python
#7.calculator using functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Choose operation: "))

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == 1:
    print("Result =", add(num1, num2))
elif choice == 2:
    print("Result =", subtract(num1, num2))
elif choice == 3:
    print("Result =", multiply(num1, num2))
elif choice == 4:
    print("Result =", divide(num1, num2))
else:
    print("Invalid Choice")
```

```python
#8.create a text file and store student details
name = input("Enter student name: ")
roll = input("Enter roll number: ")
marks = input("Enter marks: ")

with open("student.txt", "w") as file:
    file.write(f"Name: {name}\n")
    file.write(f"Roll No: {roll}\n")
    file.write(f"Marks: {marks}\n")

print("Student details saved successfully.")
```

```python
#9.read data from a file
try:
    with open("student.txt", "r") as file:
        data = file.read()
        print("File Contents:")
        print(data)

except FileNotFoundError:
    print("File does not exist.")
```

```python
#10.handle division by zero using exception handling
try:
    num1 = int(input("Enter numerator: "))
    num2 = int(input("Enter denominator: "))

    result = num1 / num2
    print("Result =", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Please enter valid numbers.")
```

```python
#11.create a student class with name and marks
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Student Name:", self.name)
        print("Marks:", self.marks)

name = input("Enter student name: ")
marks = float(input("Enter marks: "))

s1 = Student(name, marks)
s1.display()
```
